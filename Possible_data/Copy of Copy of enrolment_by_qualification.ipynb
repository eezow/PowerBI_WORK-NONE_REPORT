```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
studies = ['Undergraduates', 'Postgraduate up to masters', 'Masters and doctors']
percentages = [98, 2, 3]
ax.set_ylabel('Qualification')
ax.set_xlabel('Percentage')
ax.set_title('Percentages of enrolment by qualification')
ax.bar(studies,percentages)
plt.show()
```
